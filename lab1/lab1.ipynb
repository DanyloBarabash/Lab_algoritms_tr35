Код на мові python, який реалізує алгоритм вирішення судоку 

```python
def findNextCellToFill(grid, i, j):
        for x in range(i,9):
                for y in range(j,9):
                        if grid[x][y] == 0:
                                return x,y
        for x in range(0,9):
                for y in range(0,9):
                        if grid[x][y] == 0:
                                return x,y
        return -1,-1

def isValid(grid, i, j, e):
        rowOk = all([e != grid[i][x] for x in range(9)])
        if rowOk:
                columnOk = all([e != grid[x][j] for x in range(9)])
                if columnOk:
                        # finding the top left x,y co-ordinates of the section containing the i,j cell
                        secTopX, secTopY = 3 *(i//3), 3 *(j//3) #floored quotient should be used here. 
                        for x in range(secTopX, secTopX+3):
                                for y in range(secTopY, secTopY+3):
                                        if grid[x][y] == e:
                                                return False
                        return True
        return False

def solveSudoku(grid, i=0, j=0):
        i,j = findNextCellToFill(grid, i, j)
        if i == -1:
                return True
        for e in range(1,10):
                if isValid(grid,i,j,e):
                        grid[i][j] = e
                        if solveSudoku(grid, i, j):
                                return True
                        # Undo the current cell for backtracking
                        grid[i][j] = 0
        return False
```

Вивід готового судоку та  часу виконання алгоритму 

```python

import time
import memory_profiler 

yAxisMemory = []
yAxisTime = []
xAxis = []

i=0

while i < 50:
    xAxis.append(i)
    m1 = memory_profiler.memory_usage()
    
    start_time = time.time()
    solveSudoku(input)
    yAxisTime.append(time.time() - start_time)
    m2=memory_profiler.memory_usage() 
    yAxisMemory.append(m2[0] - m1[0])
    print(f"It took {yAxisTime[i]} Secs and {yAxisMemory[i]} Mb ")

    i=i+1
    
input

%matplotlib notebook
import matplotlib.pyplot as plt
plt.subplot(2, 2, 1)
plt.plot(xAxis, yAxisTime)

plt.subplot(2, 2, 3)
plt.plot(xAxis, yAxisMemory)

```
