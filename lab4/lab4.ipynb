```python


from random import randint
import time
import memory_profiler 
%matplotlib notebook
import matplotlib.pyplot as plt

data = [randint(0, 1000) for i in range(0, 1000)]
print(data)

workData=[]

yAxisMemory = []
yAxisTime = []


yAxisTimeBest = []
yAxisTimeWorst = []
yAxisTimeAvarage = []

best=10
worst=1000
totalTime=0





xAxis = [1,5,10]

```

```python
def partition(array, begin, end):
    pivot = begin
    for i in range(begin+1, end+1):
        if array[i] <= array[begin]:
            pivot += 1
            array[i], array[pivot] = array[pivot], array[i]
    array[pivot], array[begin] = array[begin], array[pivot]
    return pivot

def quick_sort(array, begin=0, end=None):
    if end is None:
        end = len(array) - 1
    
    def _quicksort(array, begin, end):
        if begin >= end:
            return
        pivot = partition(array, begin, end)
        _quicksort(array, begin, pivot-1)
        _quicksort(array, pivot+1, end)
    return _quicksort(array, begin, end)
```

```python
def analizeFunkTime(numOfIter):
   
    global worst
    global best
    global totalTime
    
    i=0
    totalTimeI=0
    used_time=0
    workData=data
    
    while i<10: 
        j=0
        while j<numOfIter:
            start_time = time.time()
            quick_sort(workData)
            stop_time = time.time()
            used_time+=stop_time - start_time
            j+=1 
            print (j)
            if(used_time>  best):
                worst=used_time
            else:
                best=used_time
        totalTimeI+=used_time
        i+=1   
    totalTime=totalTimeI    
    return used_time/10


def analizeFunkMb(numOfIter):
    i=0
   
    used_memory=0
    workData=data
    while i<10: 
        j=0
        while j<numOfIter:
            m1 = memory_profiler.memory_usage()
            quick_sort(workData)
            m2=memory_profiler.memory_usage() 
            used_memory+=m2[0] - m1[0]
            j+=1 
        i+=1
    return used_memory/10
```

```python
j=0
global best

for i in xAxis:
    yAxisTime.append(analizeFunkTime(i))
    yAxisMemory.append(analizeFunkMb(i))
    yAxisTimeWorst.append(worst)
    yAxisTimeBest.append(best)
    best=1000
    yAxisTimeAvarage.append(totalTime/i)   
    j+=1

    
workData=data 
quick_sort(workData)
print(workData)



 
```
